```python
# 1) Count Vowels in a String

# Task: Take a string input from the user and count how many vowels (a, e, i, o, u) it contains.

user_input = input("Count your vowel")

uppervowel = 'aeiou'.upper()
vowel = 'aeiou' + uppervowel
count = 0
for char in user_input:
    if char in vowel:
        count+=1

print(f'number of vowel is {count}')
    
```

```python
# 2) Unique Elements from Two Lists
# Task: Given two lists, create a new list containing only elements that appear in either list but not both.
# exp:
# a = [1,2,3]
# b= [1,456,2]
# output = [456, 3]

list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]
unique_elements = list(set(list1) ^ set(list2))
print(f"Unique elements: {unique_elements}")
```

```python
# 3)Word Frequency Counter
# Task: Take a sentence from user input and create a dictionary where keys are words and values are the number of times the word appears.

sentence = input("Enter a sentence: ")
words = sentence.split()

word_freq = {}
for word in words:
    word_freq[word] = word_freq.get(word, 0) + 1

print(f"Word Frequency:{word_freq}")
```

```python
# 4) Swap Keys and Values
# Task: Given a dictionary (e.g., {"a":1, "b":2}), create a new dictionary where keys become values and values become keys.
original_dict={"a":1, "b":2, "c":3}
swapped_dict = {value: key for key, value in original_dict.items()}
print(f"swapped_dict : {swapped_dict}")
```

```python
# 5) Find Maximum and Minimum Without Using max() or min()
# Task: From a list of numbers, find the largest and smallest elements manually using loops. in easy way to do it.
numbers = [34,6,12,56,7,43,13,67,24]
max_num = numbers[0]
min_num = numbers[0]
for num in numbers:
    if num > max_num:
        max_num = num
    if num < min_num:
        min_num = num
print(f'Maximum number is {max_num}')
print(f'Minimum number is {min_num}')
```

```python
# 6) Remove Duplicates from a List (Keep Order)
# Task: Take a list and return a new list with duplicates removed while maintaining order.
original_list = [1,2,3,2,4,5,1,6,3,7]
seen = set()
unique_list = []
for item in original_list:
    if item not in seen:
        seen.add(item)
        unique_list.append(item)
print(f"Unique list: {unique_list}")
```

```python
# 7.) Tuple Unpacking
# You have a tuple like (10, 20, 30, 40, 50).
# Unpack the first two variables, skip the middle values, and store the last value separately.

numbers=(10, 20, 30, 40, 50)

a,b,*_,last = numbers

print(a)    # 10
print(b)    # 20
print(last) # 50
```

```python
# 8) Merge Two Dictionaries
# Task: Write code to merge two dictionaries. If the same key exists in both, add their values.
# Example:
# {"a": 2, "b": 3} and {"a": 5, "c": 10} → {"a": 7, "b": 3, "c": 10}

dict1 = {"a":2,"b":3}
dict2 = {"a":5,"c":10}
merged_dict = dict1.copy()
for key, value in dict2.items():
    if key in merged_dict:
        merged_dict[key] += value
    else:
        merged_dict[key] = value
print(f"Merged dictionary: {merged_dict}")
```

```python
# 9) Check if a String Is a Palindrome
# Task: Check if a user-given string reads the same forwards and backwards.
user_input = input("Enter a string: ")
if user_input == user_input[::-1]:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```
